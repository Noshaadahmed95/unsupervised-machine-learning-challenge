```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import matplotlib as plt
from pathlib import Path
```

```python
file = "myopia.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
## I read the csv file inot a Pandas data frame. Now to drop the MYOPIC column so that my unsupervised model is unbiased
```

```python
labels = df['MYOPIC'].values
data = df.drop('MYOPIC', axis=1)
```

```python
data.head()
```

```python
##Sclaling the values so large values do not infuence the overall outcome
```

```python
data.describe()
```

```python
# we look at this to determine whether it will skew the model or not
```

```python
data_scaled = StandardScaler().fit_transform(data)
print(data_scaled[0:5])
```

```python
pd.DataFrame(data_scaled, columns = data.columns).describe()
```

```python
from sklearn.decomposition import PCA
```

```python
# Applying PCA to reduce dimensions

# Initialize PCA model
pca = PCA(n_components=0.99)

# Get two principal components for the iris data.
data_pca = pca.fit_transform(data_scaled)
data_pca
```

```python
data_pca.shape
```

```python
# Transform PCA data to a DataFrame
df_data_pca = pd.DataFrame(
    data=data_pca, columns=["component1", "component_2", "component_3", "component_4", "component_5", "component_6", "component_7", "component_8", "component_9", "component_10", "component_11", "component_12"]
)
df_data_pca.head()
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(learning_rate=35)
```

```python
tsne_features = tsne.fit_transform(data_pca)
```

```python
tsne_df = pd.DataFrame(tsne_features)
```

```python
tsne_df
```

```python
tsne_df = tsne_df.rename(columns={0: "X", 1: "Y"})
tsne_df
```

```python
import matplotlib.pyplot as plt
```

```python
# Prepare to plot the dataset

plt.scatter(tsne_df.X, tsne_df.Y)
```

```python
 # Results colored by target value
plt.scatter(tsne_df.X, tsne_df.Y, c=labels)

plt.show()
```

```python
## Perform a Cluster Analysis with K-means
from sklearn.cluster import KMeans
import numpy as np
```

```python
for column in tsne_df.columns:
    print(f"{column} = {tsne_df[column].isnull().sum()}")
```

```python
 # Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(tsne_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Predicting clusters with k=3

# Initialize the k-means model
model = KMeans(n_clusters=3, random_state=0)

# Fit the model
model.fit(tsne_df)

# Predict clusters
predictions = model.predict(tsne_df)

# Add the predicted class column to the dataframe
tsne_df["class"] = model.labels_
tsne_df.head()
```
